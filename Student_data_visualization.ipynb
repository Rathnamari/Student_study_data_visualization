```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ELCOT\Downloads\archive\student_performance_prediction.csv")
df
```

```python
df.info()
```

```python
df['Parent Education Level'].fillna(value = "uneducated")
df['Study Hours per Week'].fillna(value = df['Study Hours per Week'].mean())
df['Attendance Rate'].fillna(value = df['Attendance Rate'].mean())
df['Previous Grades'].fillna(value = df['Previous Grades'].mean())
df['Participation in Extracurricular Activities'].fillna( method = 'ffill')
df['Passed'].fillna( method = 'ffill')
```

```python
df.info()
```

```python
df['Parent Education Level'].unique()
```

```python
# Concatenate all values in the 'Parent Education Level' column into a single string
text = ','.join(df["Parent Education Level"].dropna())
text1 = text.split(',')
```

```python
unique_textss = []
unique_text = set(text1)
for i in unique_text:
    if i == 'High School':
        unique_textss.append('HighSchool')
    else:
        unique_textss.append(i)
unique_textss
```

```python
unique_texts = ' '.join(unique_textss)
unique_texts
```

```python
wc = WordCloud(background_color = "black", width = 1440, height = 1080, colormap = 'plasma').generate(unique_texts)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
